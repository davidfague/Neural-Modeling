```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the DataFrame is named df and it is already loaded
# df = pd.read_csv('your_file.csv')  # Load your DataFrame from a CSV file

# Example DataFrame based on your provided structure
data = {
    "Unnamed: 0": [0, 1, 2, 3, 4, 5, 6, 7],
    "cell": ["Complex", "ComplexMappingMerging", "Branches", "BranchesMappingMerging", "BranchesREPBasals", "BranchesREPBasalsMappingMerging", "Trees", "TreesMappingMerging"],
    "runtime": [885.6, 895.0, 130.0, 126.3, 100.2, 95.7, 44.6, 42.8],
    # other columns...
}

df = pd.DataFrame(data)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['cell'], df['runtime'], color='skyblue')
plt.xlabel('Cell')
plt.ylabel('Runtime')
plt.title('Runtime for Each Cell')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the plot
plt.show()
```
