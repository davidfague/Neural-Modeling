```python
def compute_distance_from_one(value):
    return abs(value - 1)

def compute_percent_change(new_value, old_value):
    return ((new_value - old_value) / old_value ) * 100

def show_diff(stats, old_model='BranchesMappingMerging', new_model = 'Branches'):
    for spike_relationship in stats[old_model]:
        print(spike_relationship, average_mses[old_model][spike_relationship], average_mses[new_model][spike_relationship])
        
def show_percent_diff(stats, old_model='BranchesMappingMerging', new_model = 'Branches'):
    for spike_relationship in stats[old_model]:
        print(spike_relationship, compute_percent_change(average_mses[old_model][spike_relationship], average_mses[new_model][spike_relationship]))

def show_percent_diff_dist_from_1(stats, old_model='BranchesMappingMerging', new_model = 'Branches'):
    for spike_relationship in stats[old_model]:
        print(spike_relationship, compute_percent_change(compute_distance_from_one(stats[old_model][spike_relationship]), compute_distance_from_one(stats[new_model][spike_relationship])))
```

```python
# R2
show_percent_diff_dist_from_1(average_r_squareds, old_model='Complex', new_model = 'Branches')
```

```python
# MSES
show_diff(average_mses, old_model='Complex', new_model = 'Branches')
```

```python
# Function to sum columns and count NaNs
def sum_and_count_nans(row, columns):
    values = row[columns]
    sum_val = values.sum(skipna=True)
    nan_count = values.isna().sum()
    formatted_sum = f"{sum_val:.2e}"
    return (formatted_sum, nan_count)

def mean_and_count_nans(row, columns):
    values = row[columns]
    mean_val = values.mean(skipna=True)
    nan_count = values.isna().sum()
    formatted_mean = f"{mean_val:.2e}"
    return (formatted_mean, nan_count)

# Identify columns that start with 'average_mae_'
average_mae_columns = [col for col in df.columns if col.startswith('average_mae_')]
# Calculate the tuple (sum, NaN count) for each row
df['all_spikes_mae'] = df.apply(lambda row: mean_and_count_nans(row, average_mae_columns), axis=1)

# Insert the new column at the desired position (between 3rd and 4th columns)
cols = df.columns.tolist()
position = 3  # Insert after the current 3rd column
cols.insert(position, cols.pop(cols.index('all_spikes_mae')))
df = df[cols]

# Identify columns that start with 'average_mse_'
average_mse_columns = [col for col in df.columns if col.startswith('average_mse_')]
# Calculate the tuple (sum, NaN count) for each row
df['all_spikes_mse'] = df.apply(lambda row: mean_and_count_nans(row, average_mse_columns), axis=1)

# Insert the new column at the desired position (between 3rd and 4th columns)
cols = df.columns.tolist()
position = 3  # Insert after the current 3rd column
cols.insert(position, cols.pop(cols.index('all_spikes_mse')))
df = df[cols]

# Identify columns that start with 'average_r2_'
average_r2_columns = [col for col in df.columns if col.startswith('average_r2_')]
# Calculate the tuple (sum, NaN count) for each row
df['all_spikes_r2'] = df.apply(lambda row: mean_and_count_nans(row, average_r2_columns), axis=1)

# Insert the new column at the desired position (between 3rd and 4th columns)
cols = df.columns.tolist()
position = 3  # Insert after the current 3rd column
cols.insert(position, cols.pop(cols.index('all_spikes_r2')))
df = df[cols]

# Display the updated DataFrame
# print(df)
```
